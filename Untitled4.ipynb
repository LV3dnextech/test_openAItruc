```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd # Assicurati che pandas sia importato

# Assicurati che il DataFrame 'df_completed' sia disponibile.
if df_completed is not None:
    # Per visualizzare la densità, possiamo utilizzare un grafico a dispersione dove il colore
    # dei punti rappresenta la densità dei dati sottostanti.

    # Seleziona le colonne di interesse
    x = df_completed['rifinitura_minutes']
    y = df_completed['t_med']

    # Calcola la densità dei punti (utile per colorare)
    # Utilizziamo una griglia 2D per contare i punti in ciascuna area
    try:
        # Rimuovi i valori mancanti se ce ne fossero stati (anche se dovrebbero essere stati gestiti)
        data = df_completed[['rifinitura_minutes', 't_med']].dropna()

        # Usa numpy.histogram2d per calcolare la densità
        hist, xedges, yedges = np.histogram2d(data['rifinitura_minutes'], data['t_med'], bins=[50, 50])
        # Normalizza la densità per ottenere percentuali o una scala di colore
        hist_norm = hist / hist.sum() * 100 # Percentuale

        # Crea una lista di colori in base alla densità di ciascun punto originale
        # Questo è un po' più complesso e non rappresenta la densità esatta di ogni singolo punto
        # Un approccio più comune è usare un contour plot o un density plot,
        # ma se vogliamo colorare i punti, possiamo mappare la densità locale.
        # Un modo più semplice per colorare basandosi sulla "densità" percepita è usare un cmap.

        # Un approccio più diretto per colorare i punti in base alla densità
        # è calcolare la densità del kernel o utilizzare hexbin per aree del grafico
        # Per un grafico a dispersione che "sembra" mostrare la densità,
        # possiamo colorare i punti in base alla loro prossimità ad altri punti (non ideale)
        # o semplicemente usare un alpha value (trasparenza) o dimensioni diverse per i punti.

        # Un approccio visivo semplice per suggerire la densità in uno scatter plot
        # è usare la trasparenza (alpha) o la dimensione dei punti (size).
        # Un altro modo è colorare in base alla densità calcolata sui bin.
        # Per semplicità e visualizzazione chiara, proviamo a colorare i punti in base
        # alla densità del bin in cui si trovano.

        # Mappa ogni punto al suo bin e ottieni la densità di quel bin
        # Questo richiede di trovare l'indice del bin per ogni punto.
        # Utilizzeremo digitize per trovare l'indice del bin per ogni punto
        x_indices = np.digitize(data['rifinitura_minutes'], xedges) - 1
        y_indices = np.digitize(data['t_med'], yedges) - 1

        # Assicurati che gli indici siano all'interno dei limiti dei bin
        x_indices[x_indices == len(xedges) -1] = len(xedges) - 2
        y_indices[y_indices == len(yedges) -1] = len(yedges) - 2
        x_indices[x_indices < 0] = 0
        y_indices[y_indices < 0] = 0


        # Ottieni la densità normalizzata per ciascun punto in base al suo bin
        point_density = hist_norm[x_indices, y_indices]


        # Crea il grafico a dispersione
        plt.figure(figsize=(12, 7))

        # Usiamo lo scatter plot colorando i punti in base alla densità calcolata
        scatter = plt.scatter(data['rifinitura_minutes'], data['t_med'], c=point_density, cmap='viridis', s=10) # s controlla la dimensione dei punti
        plt.colorbar(scatter, label="Densità (% di lavorazioni in questo bin)") # Aggiungi una barra dei colori

        plt.xlabel("Tempo di Rifinitura (minuti)")
        plt.ylabel("Temperatura Media (°C)")
        plt.title("Relazione tra Tempo di Rifinitura e Temperatura Media (Lavorazioni Completate)")
        plt.grid(True, linestyle='--', alpha=0.6)
        plt.show()

    except Exception as e:
        print(f"Errore nella creazione del grafico di densità: {e}")
        print("Potrebbe esserci un problema con i dati o con il calcolo della densità.")


else:
    print("DataFrame 'df_completed' non trovato. Si prega di assicurarsi che sia stato creato.")
```
